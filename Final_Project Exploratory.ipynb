```python
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
# Handling missing values
for column in data.columns:
    if data[column].dtype == 'float64' or data[column].dtype == 'int64':
        data[column].fillna(data[column].mean(), inplace=True)
    else:
        data[column].fillna(data[column].mode()[0], inplace=True)
# Encoding categorical variables
label_encoder = LabelEncoder()
categorical_columns = ['Geography', 'Gender']
for column in categorical_columns:
    data[column] = label_encoder.fit_transform(data[column])
# Normalizing numerical variables
numerical_columns = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts']
scaler = StandardScaler()
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


# Analyzing distributions of features
plt.figure(figsize=(12, 8))
for i, column in enumerate(['Age', 'CreditScore', 'Balance', 'EstimatedSalary'], start=1):
    plt.subplot(2, 2, i)
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()

# Drop non-numeric columns to avoid errors in correlation matrix calculation
numeric_data = data.select_dtypes(include=[np.number])

# Examine relationships between numeric variables
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='vlag')
plt.title('Correlation Matrix')
plt.show()


# Compute the correlation matrix
corr_matrix = numeric_data.corr()
filtered_corr = corr_matrix[(corr_matrix >= 0.20) | (corr_matrix <= -0.20)]

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))
sns.heatmap(filtered_corr, annot=True, cmap='vlag', center=0)
plt.title('Filtered Correlation Matrix')
plt.show()

#Exited Vs. Age
plt.figure(figsize=(6, 4))
sns.boxplot(x='Exited', y='Age', data=numeric_data)
plt.title('Age vs Exited')
plt.show()
```
