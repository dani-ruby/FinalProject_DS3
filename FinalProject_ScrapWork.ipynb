```python
import pandas as pd
```

```python
data = pd.read_csv('/Users/daniloruberti/Desktop/train.csv')
```

```python
data.tail()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


# Analyzing distributions of features
plt.figure(figsize=(12, 8))
for i, column in enumerate(['Age', 'CreditScore', 'Balance', 'EstimatedSalary'], start=1):
    plt.subplot(2, 2, i)
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()

# Drop non-numeric columns to avoid errors in correlation matrix calculation
numeric_data = data.select_dtypes(include=[np.number])

# Examine relationships between numeric variables
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='vlag')
plt.title('Correlation Matrix')
plt.show()

#Exited Vs. Age
plt.figure(figsize=(6, 4))
sns.boxplot(x='Exited', y='Age', data=numeric_data)
plt.title('Age vs Exited')
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


for column in ['Age', 'CreditScore', 'Balance', 'EstimatedSalary']:
    plt.figure(figsize=(12, 8)) 
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.tight_layout()
    plt.show() 
# Drop non-numeric columns to avoid errors in correlation matrix calculation
numeric_data = data.select_dtypes(include=[np.number])

# Examine relationships between numeric variables
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='vlag')
plt.title('Correlation Matrix')
plt.tight_layout()  # It's good practice to use tight_layout before showing the plot
plt.show()

# Exited Vs. Age
plt.figure(figsize=(6, 4))
sns.boxplot(x='Exited', y='Age', data=numeric_data)
plt.title('Age vs Exited')
plt.tight_layout()  
plt.show()
```
